```python
import pandas as pd
calendardf = pd.read_table('calendar.txt',sep=',')          #export the calendar file into data frame
calendardf.head(30)
```

```python
#Need to create a calendar for august and september

#We create a dict such date the key is the format 'MonthDay' (exemple '827' correspond to the 27th of August)
# and the value is the day in the week

DayOfWeek = ['monday','tuesday','wednesday','thursday','friday','saturday','sunday']

calendarPierre={}

for i in range(31):    #loop for August
    date=801+i
    date=str(date)
    calendarPierre[date]=DayOfWeek[(i+6)%7]         #The 1th of August is a Sunday

for j in range(30):         #loop for September
    date=901+j
    date=str(date)
    calendarPierre[date]=DayOfWeek[(j+2)%7]         #The 1th of Sept is a Tuesday
```

```python
def FindDays(serviceId):

    #the function takes a serviceId in argument and return the days of validity it includes, in the format 'MonthDay'

    dataframe=calendardf[calendardf['service_id']==serviceId]    

    #We can get rid of the year here 

    startdate=dataframe['start_date'].values-20210000         
    enddate=dataframe['end_date'].values-20210000
    DaysOfValidity=[]

    for day in range(int(startdate),int(enddate)+1):
        day = str(day)
        if dataframe[calendarPierre[day]].values==1:
            DaysOfValidity.append({str(day):calendarPierre[day]})
            
    return(DaysOfValidity)
```

```python
def CheckRange(Start,End,date):
    return int(date) in range(int(Start),int(End))

def FromDayToService(date):
    #date est un tuple ('20210907','tuesday')
    ListServiceId=[]

    for i in range(len(calendardf)):

        startdate,endate=calendardf.iloc[i,[8,9]]
        
        if CheckRange(startdate,endate+1,date[0]):

            if calendardf.loc[i,date[1]]==1:
                
                ListServiceId.append(str(calendardf.iloc[i,0]))

    return ListServiceId
FromDayToService(('20210907','tuesday'))

```
